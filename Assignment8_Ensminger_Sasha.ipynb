**Problem 1 (3 points)**

We can generalize the Fibonacci sequence so that we have the recurrence relationship

$$
p_{n} = ap_{n-1} + bp_{n-2}, ~ b \in \mathbb{R}, ~ n\geq 2,
$$

where we start the sequence $p_{n}$ with the initial conditions

$$
p_{0} = s_{0}, ~ p_{1} = 1, ~ s_{0} \in \mathbb{R}.
$$

[a] (3pts) Using the guess $p_{n} = \lambda^{n}$, show that you get two solutions for $\lambda$, say $\lambda_{\pm}$, where

$$
\lambda_{\pm} = \frac{1}{2}\left(a \pm \sqrt{a^2+4b}\right)
$$

[b] (3pts) Writing the general solution as 

$$
p_{n} = c_{+}\lambda_{+}^{n} + c_{-}\lambda_{-}^{n}
$$

show that when we take our initial conditions into account, we find that the constants $c_{+}$ and $c_{-}$ are given by 

$$
c_{+} = \frac{1-s_{0}\lambda_{-}}{\lambda_{+}-\lambda_{-}}, ~ c_{-} = \frac{s_{0}\lambda_{+}-1}{\lambda_{+}-\lambda_{-}}
$$

[c] (4pts) Let $b = 1$ and $a=2$.  

i.) Determine which of the following is correct (you must show work to obtain full credit): 
$$
A.) ~ \left|\lambda_{+}\right| > 1, \quad B.) \left|\lambda_{+} \right|<1
$$ 


ii.) Determine which of the following is correct (you must show work to obtain full credit): 
$$
A.) ~ \left|\lambda_{-}\right| > 1, \quad B.) \left|\lambda_{-} \right|<1
$$

### Solution for Problem 1a ###

Since $p_{n} = ap_{n-1} + bp_{n-2}$  and  $p_{n} = \lambda^{n}$, <br>
$$
\lambda^{n} = \lambda^{n-1} + b\lambda^{n-2} \\
\lambda \neq 0  ~ ~ \text{and} ~ ~ \lambda^{n-2} \neq 0 \\
\lambda^{2} - a\lambda - b = 0 \\
\lambda_{\pm} = \frac{-(-a) \pm \sqrt{(-a)^{2}-4(1)(-1)}}{2(1)} \\
\lambda_{+} = \frac{a + \sqrt{a^{2}+4b}}{2} ~ ~ \text{and} ~ ~ \lambda_{-} = \frac{a - \sqrt{a^{2}+4b}}{2}
$$

### Solution for Problem 1b ###

Since $p_{0} = s_{0}$  and  $p_{1} = 1$, <br>
$$
s_{0} = c_{+} + c_{-} \text{ - - - - {1}} \\
1 = c_{+}\lambda_{+} + c_{-}\lambda_{-} \text{ - - - - {2}} \\
\text{{1}}\cdot\lambda_{+} + \text{{2}}\cdot1 \\
\lambda_{+}s_{0} = c_{+}\lambda_{+} + c_{-}\lambda_{-} \text{ - - - - {3}}\\
1 = c_{+}\lambda_{+} + c_{-}\lambda_{-} \text{ - - - - {4}}\\
$$ 
Combine equations {3} and {4},
$$
\lambda_{+}s_{0}-1 = c_{-}(\lambda_{+}-\lambda{-}) \\
c_{-} = \frac{\lambda_{+}s_{0}-1}{\lambda_{+}-\lambda{-}}
$$
Putting this value into {2},
$$
\begin{align}
c_{+} = & s_{0} - \frac{\lambda_{+}s_{0}-1}{\lambda_{+}-\lambda{-}} \\
= & \frac{s_{0}\lambda_{+}-s_{0}\lambda_{-}-s_{0}\lambda_{+}+1}{\lambda_{+}-\lambda_{-}} \\
= & \frac{1-s_{0}\lambda_{-}}{\lambda{+}-\lambda{-}}
\end{align}
$$

### Solution for Problem 1c ###

Since $b = 1$ and $a = 2$,
$$
\begin{align}
\lambda_{+} = & \frac{2 + \sqrt{4+4}}{2} \\
= & \frac{2 + \sqrt{8}}{2} \\
= & 1 + \sqrt{2}
\lambda_{-} = & \frac{2 - \sqrt{4+4}}{2} \\
= & \frac{2 - \sqrt{8}}{2} \\
= & 1 - \sqrt{2}
\end{align}
$$
<br>
$|\lambda_{+}| = |1 + \sqrt{2}| > 1 ~ ~ \Rightarrow ~ ~ $ Answer A  
$|\lambda_{-}| = |1 - \sqrt{2}| < 1 ~ ~ \Rightarrow ~ ~ $ Answer B   


**Problem 2 (6 points)**

The $3n+1$ sequence is generated using the following rules:

- Start with a positive integer n.

- If n = 1, stop.

- If n is even, replace it with n/2.

- If n is odd, replace it with 3n+1.

So if we started with $n=3$, then we would generate the sequence

$$
3, 10, 5, 16, 8, 4, 2, 1
$$

The unsolved mathematical problem is whether this code can run forever i.e. it is unknown whether there are any starting values $n$ which generate a sequence which goes on for forever.  Thus, an interesting associated quantity we would want to know is how many terms a given value of $n$ generates via the $3n+1$ sequence.  We call this number $L(n)$.  For example then, using our example of $n=3$ above, we have that 

$$
L(3)  = 8
$$

Write a program to find $L(n)$ for any given $n$ and then generate a plot of $L(n)$ for $1\leq n \leq 1000$.  Discuss any trends you observe.  You may modify the code I gave in class to accomplish the task or write your own.

```python
#Solution for Problem 2

def L(n):
    count = 1
    while n > 0:
        if n == 1:
            break
        elif n%2 == 0: 
            n = n/2
        else: 
            n = 3*n + 1
        count += 1
    return count
```

**Problem 4 (3 points)**

a) A sequence {$a_n$} is defined recursively by the equations:

$$a_0 = a_1 = 1$$

$$n(n-1)a_n = (n-1)(n-2)a_{n-1} - (n-3)a_{n-2}.$$

Write a program to determine the following sum:

$$\sum_{n=0}^{1001} a_n$$

b) Run your code and provide the answer for (4a)

```python
#Solution for Problem 4

def summ(n):
    if n == 0 or n == 1: 
        return 1
    term = ((n-2)*summ(n-1))/n - (((n-3)*summ(n-2))/(n*(n-1)))
    return term + summ(n-1)
```

$$

[a] Using the skeleton-code below write a function `Lucas` which generates, for given integer value $n\geq 2$ and real values `a`,`b`, `s0`, and `s1`, the $n+1$ array of points `pn` such that 

`
pn[0] = s0, pn[1] = s1
`

and the entries in `pn` are given by the recurrence relationship from Problem 1 above.

[b] Fixing $n=15$, $s_{0}=0$, $s_{1}=2$, $a=2$, $b=1$  generate a plot of $p_{n}$.  Make sure axes are appropriately labeled.  

[c] Fixing $n=10$, $s_{0}=1$, $s_{1}=1$, $a=1$ choose $b=-1/2$ and generate a plot of $p_{n}$.  Briefly explain the results you see and how they differ from those in [b]

```python
# Solution for Problem 6a

def Lucas(n, a, b, s0, s1):
    pn = [s0, s1]
    for ii in range(2, n+1):
        pn.append(a*pn[ii-1] + b*pn[ii-2])
    return pn
```
